```python
from keras.models import load_model
from collections import deque 
import numpy as np
import pickle
import math
import cv2
```

```python
CATEGORIES = ['Shoplifting', 'Normal Activity']

outputmodelpath = r'model/vgg16'

path = r"model/vgg16/"
bi_path = r"model/vgg16/binarizer.pkl"
output_video = r"../Data/output/"
model = load_model(path)

with (open(bi_path, "rb")) as openfile:
    while True:
        try:
            lb = pickle.load(openfile)
        except EOFError:
            break

queue = deque(maxlen=128)
```

```python
# vid_path = r"../Data/Shoplifting/video37.mp4"
vid_path = r"../Data/Fight/Fight_OneManDown.mpg"

capture_video = cv2.VideoCapture(vid_path)
writer = None
(Width, Heigth) = (None, None)

while True:
    (taken, frame) = capture_video.read()
    if not taken:
        break
    if Width is None or Height is None:
        (Width, Height) = frame.shape[:2]
        
    output = frame.copy()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frame = cv2.resize(frame, (224, 224)).astype("float32")
        
    preds = model.predict(np.expand_dims(frame, axis=0))[0]
    results = np.array(queue).mean(axis=0)
    i = int(round(preds[0]))
    queue.append(i)
    print(preds, "", i)
    label = lb.classes_[i]
    text = "{}".format(label)
    cv2.putText(output, text, (45, 60), cv2.FONT_HERSHEY_SIMPLEX, 1.25, (255,0,0), 5)
    
    if writer is None:
        fourcc = cv2.VideoWriter_fourcc(*"MJPG")
        writer = cv2.VideoWriter("output_video", fourcc, 30, (Width, Height), True)
    writer.write(output)
    cv2.imshow("In progress", output)
    key = cv2.waitKey(1) & 0xFF
    
    if key == ord("q"):
        break
        
print("Finalizing....")
writer.release()
capture_video.release()
cv2.destroyAllWindows()

print(sum(queue) / len(queue))
```

```python
from sklearn import metrics

# Printing the confusion matrix
# The columns will show the instances predicted for each label,
# and the rows will show the actual number of instances for each label.
print(metrics.confusion_matrix(y_actual, y_predicted, labels=[0, 1]))
# Printing the precision and recall, among other metrics
print(metrics.classification_report(y_actual, y_predicted, labels=[0, 1]))
```
